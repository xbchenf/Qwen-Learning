# 启动本地 Web UI 演示
官方为用户提供构建基于 Web 的用户界面（UI）演示的说明。此 UI 演示允许用户通过 Web 浏览器与预定义的模型或应用程序进行交互。

# 一、安装相关依赖

在开始之前，请确保已将所需的依赖项安装到系统上。我们可以通过运行以下命令来安装它们：
```bash
git clone https://github.com/QwenLM/Qwen2.5-Omni.git
cd Qwen2.5-Omni
pip install -r requirements_web_demo.txt
```


# 二、运行启动web服务
打开根目录下web_demo.py文件修改模型地址（将远程仓库地址改为本地下载的模型权重文件地址）

修改前：
DEFAULT_CKPT_PATH = "Qwen/Qwen2.5-Omni-7B"

修改后：
DEFAULT_CKPT_PATH = "/root/autodl-tmp/Qwen/Qwen2.5-Omni-7B"

# 三、运行启动web服务

推荐：为了提高性能和效率，尤其是在多图像和视频处理场景中，强烈建议使用 FlashAttention-2。FlashAttention-2 在内存使用和速度方面提供了显著的改进，非常适合处理大规模模型和数据。

## 1）启用 FlashAttention-2
请使用以下命令：
```bash
python web_demo.py --flash-attn2
```
启动完成如下：


## 2）默认用法（不启用 FlashAttention-2）：
如果希望在不使用 FlashAttention-2 的情况下运行演示，或者没有指定 --flash-attn2 选项，则演示将使用标准注意力实现加载模型：
```bash
python web_demo.py
```


# 四、访问UI对话界面
运行命令后，您将在终端中看到生成的链接，类似于以下内容：
Running on local: http://127.0.0.1:7860/
复制此链接并将其粘贴到浏览器中，以访问 Web UI，您可以在其中通过输入文本、上传音频/图片/视频、更改声音类型或使用提供的任何其他功能与模型进行交互。
