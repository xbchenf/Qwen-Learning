
```bash

```

# 使用示例

````python
import subprocess

# 运行 Qwen3 模型并获取回答
def get_response(prompt, model="qwen3:8b"):
    command = f"echo '{prompt}' | ollama run {model}"
    result = subprocess.run(command, shell=True, capture_output=True, text=True)
    return result.stdout

# 示例：获取关于大型语言模型的简短介绍
prompt = "请给我简单介绍一下大模型"
response = get_response(prompt)
print("Response:", response)
````

# 高级使用示例

````python
import subprocess
import os

# 设置模型推理参数
os.environ["OLLAMA_MAX_LENGTH"] = "100"
os.environ["OLLAMA_TEMPERATURE"] = "0.7"

# 运行 Qwen3 模型并获取回答
def get_response(prompt, model="qwen3:8b"):
    command = f"echo '{prompt}' | ollama run {model}"
    result = subprocess.run(command, shell=True, capture_output=True, text=True)
    return result.stdout

# 示例：获取关于人工智能的未来发展趋势的详细分析
prompt = "What are the future trends of artificial intelligence?"
response = get_response(prompt)
print("Response:", response)
````
